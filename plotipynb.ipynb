```python
import pandas as pd
import csv
import numpy as np
import matplotlib as mpl
from matplotlib import style 
from matplotlib import pyplot as plt

style.use('default')

x1,Eg_AlGaAs,alc_AlGaAs,so_AlGaAs,mhh_AlGaAs,mlh_AlGaAs,me_AlGaAs,VBO_AlGaAs = np.loadtxt('AlGaAs.csv', unpack=True, delimiter=',')
x2,Eg_AlAsSb,alc_AlAsSb,so_AlAsSb,mhh_AlAsSb,mlh_AlAsSb,me_AlAsSb,VBO_AlAsSb = np.loadtxt('AlAsSb.csv', unpack=True, delimiter=',')
x3,Eg_AlGaSb,alc_AlGaSb,so_AlGaSb,mhh_AlGaSb,mlh_AlGaSb,me_AlGaSb,VBO_AlGaSb = np.loadtxt('AlGaSb.csv', unpack=True, delimiter=',')
x4,Eg_GaAsSb,alc_GaAsSb,so_GaAsSb,mhh_GaAsSb,mlh_GaAsSb,me_GaAsSb,VBO_GaAsSb = np.loadtxt('GaAsSb.csv', unpack=True, delimiter=',')
```

```python
plt.plot(alc_AlGaAs, Eg_AlGaAs, label= 'AlGaAs' )
plt.plot(alc_AlAsSb, Eg_AlAsSb, label= 'AlAsSb')
plt.plot(alc_AlGaSb, Eg_AlGaSb, label= 'AlGaSb')
plt.plot(alc_GaAsSb, Eg_GaAsSb, label= 'GaAsSb')
plt.xlabel('Parametr sieci $a_{lc}$ (A)')
plt.ylabel('Energia (eV)')
plt.title('Wykres zminay szerokości przerwy energetycznej\nwzależności od parametru sieci a dla materiałów trójskładnikowych')
plt.legend(loc='best')
```
